```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
cancer= load_breast_cancer()
X= cancer.data
y= cancer.target
print(X.shape)
print(y.shape)

X= pd.DataFrame(X)
y= pd.DataFrame(y)
data= X
data.head(2)

```

```python
data.columns= cancer.feature_names
```

```python
data.head(2)
```

# **1. Feature Selection by Percentile (chi2 , f_classif) **

```python
from sklearn.feature_selection import SelectPercentile, chi2 , f_classif 
FeatureSelection = SelectPercentile(score_func = chi2, percentile=50) # score_func can be ( f_classif )
data= FeatureSelection.fit_transform(data, y)
print(data.shape)
data= pd.DataFrame(data)
data.head()
```

```python
releveant_features= FeatureSelection.get_support(indices=True)
releveant_features
```
